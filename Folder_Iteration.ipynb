```python
import os
import rasterio as rio
import matplotlib.pyplot as plt
import math

def iterate_files_in_folder(folder_path, extension):
    file_paths = []
    # Iterate over files in the folder
    for file_name in os.listdir(folder_path):
        # Check if the file has the specified extension
        if file_name.endswith(extension):
            file_path = os.path.join(folder_path, file_name)
            file_paths.append(file_path)
    return file_paths

# Example usage:
folder_path = "F:\Project\layerstack"
extension = '.img'  # Specify the extension you want to filter
files_with_extension = iterate_files_in_folder(folder_path, extension)
print("Files with extension {}: {}".format(extension, files_with_extension))
```

```python
files_with_extension
```

```python
# Loop over the list of file names
for idx, file_name in enumerate(files_with_extension, 1):  # Start index from 1
    # Open the TIFF file
    with rio.open(file_name) as src:
        # Read the raster data
        raster_data = src.read()

    # Plot the raster data
    plt.figure(figsize=(8, 8))
    plt.imshow(raster_data[0], cmap='viridis')  # Assuming the raster data is a single-band image
    plt.title(f"Raster Plot: {file_name}")
    plt.xlabel('Column')
    plt.ylabel('Row')
    plt.colorbar(label='Pixel Value')
    plt.grid(True)
    
    # Save the image with proper numbering
    plt.savefig(f"F:\\Project\\layerstack\\Image_Iter\\image_{idx}.png")
    
    # Close the plot to avoid displaying the plot multiple times
    plt.close()

# Display a message indicating that the images have been saved
print("All images have been saved successfully.")
```

# For SubPlots

```python
# Determine the number of rows and columns for the subplot grid
num_files = len(files_with_extension)
num_rows = math.ceil(num_files / 2)

# Loop over the list of file names
for idx, file_name in enumerate(files_with_extension, 1):  # Start index from 1
    # Open the TIFF file
    with rio.open(file_name) as src:
        # Read the raster data
        raster_data = src.read()

    # Plot the raster data
    plt.figure(figsize=(8, 8))
    plt.subplot(num_rows, 2, idx)  # Create subplot
    plt.imshow(raster_data[0], cmap='nipy_spectral')  # Assuming the raster data is a single-band image
    plt.title(f"Image {idx}")  # Label the subplot
    plt.xlabel('Column')
    plt.ylabel('Row')
    plt.colorbar(label='Pixel Value')
    plt.grid(True)
#        Save the image with proper numbering
#     plt.savefig(f"F:\\Project\\layerstack\\Image_Iter\\image_{idx}.png")
# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
